```python
app_train.head()
# count: 資料數
# mean: 平均值
# std: 標準差
# min: 最小值
# 25%: 25% 資料比例落在
# 50%: 50% 資料比例落在
# 75%: 75% 資料比例落在
# max: 最大值
app_train.describe()
```

```python
# 計算收入的平均數與標準差
print("AMT_ANNUITY 平均數 : " + str(app_train['AMT_ANNUITY'].mean()))
print("AMT_ANNUITY 標準差 : " + str(app_train['AMT_ANNUITY'].std()))
```
