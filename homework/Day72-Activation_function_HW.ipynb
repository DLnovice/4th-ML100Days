# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def relu(x):
    return abs(x) * (x > 0)
def div_relu(x):
    return 1 * (x > 0)

x = np.linspace(-10, 10, 100)
plt.plot(x, relu(x), 'b', label = 'relu')
plt.plot(x, div_relu(x), 'r', label = 'div_relu')
plt.legend('best')
plt.xlabel('x')
plt.ylabel('y')
```
