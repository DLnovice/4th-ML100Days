```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
# 顯示出資料共有(X行,Y列)

print('資料 row 數: ' + str(app_train.shape[0]))
print('資料 column 數: ' + str(app_train.shape[1]))
```

#### 列出所有欄位

```python
print(app_train.columns)
```

#### 截取部分資料

```python
# 選第幾筆資料的某些欄位 用loc（location的意思）中括號裡面第一個放index的範圍，第二個放column的名稱

app_train.loc[5:10,['CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY','AMT_INCOME_TOTAL','OCCUPATION_TYPE']]
```

#### 看有幾筆資料、每個欄位的資料型別是什麼(int, float..)、有無空值(null)的存在、佔據多少記憶體

```python
app_train.info()
```

#### 資料篩選  在中括號裡面方入篩選條件即可

```python
app_train[app_train['AMT_INCOME_TOTAL']>00000.0]
```

#### 資料分群 (Data Grouping)  
#### 使用groupby()方法可以將資料依照自己要的column分組

```python
# 使用 CODE_GENDER 用性別來分組
CODE_GENDER = app_train.groupby("CODE_GENDER")
# 選男人 M
CODE_GENDER.get_group("F")

#app_train.groupby(by='AMT_INCOME_TOTAL').sum()
```
