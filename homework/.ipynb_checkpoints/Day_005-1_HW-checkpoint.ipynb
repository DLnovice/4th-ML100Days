```python
import pandas as pd
import numpy as np
import random
```

```python
# 建立8個人口亂數
population = np.random.randint(10000,size = 8)

data = {'國家':['台灣','日本','韓國','美國','英國','法國','德國','西班牙'] ,
        '人口':[population[0],population[1],population[2],population[3],population[4],population[5],population[6],population[7]] }
data = pd.DataFrame(data)
print(data)
```

```python
# 取得最大值的索引
#Idx = data['人口'].idxmax()
#print(Idx)

print(data.max(0))
```
